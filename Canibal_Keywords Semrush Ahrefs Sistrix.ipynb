```python
# Instalar las bibliotecas necesarias
!pip install pandas nltk==3.6.7 openpyxl transformers sentence-transformers tqdm

# Limpiar la consola
from IPython.display import clear_output
clear_output()

# Importar las bibliotecas
import os
import pandas as pd
import nltk
from nltk.corpus import wordnet
from sentence_transformers import SentenceTransformer, util
from tqdm import tqdm
from google.colab import files

# Descargar los recursos necesarios manualmente
import urllib.request
import zipfile
import os

nltk_data_dir = os.path.expanduser('~/nltk_data')
if not os.path.exists(nltk_data_dir):
    os.makedirs(nltk_data_dir)

def download_and_extract(url, dest):
    zip_path, _ = urllib.request.urlretrieve(url)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(dest)

download_and_extract('https://raw.githubusercontent.com/nltk/nltk_data/gh-pages/packages/corpora/wordnet.zip', nltk_data_dir + '/corpora')
download_and_extract('https://raw.githubusercontent.com/nltk/nltk_data/gh-pages/packages/corpora/omw-1.4.zip', nltk_data_dir + '/corpora')
download_and_extract('https://raw.githubusercontent.com/nltk/nltk_data/gh-pages/packages/tokenizers/punkt.zip', nltk_data_dir + '/tokenizers')
download_and_extract('https://raw.githubusercontent.com/nltk/nltk_data/gh-pages/packages/taggers/averaged_perceptron_tagger.zip', nltk_data_dir + '/taggers')

# Mensaje de inicio y agradecimiento
print("‚ö†Ô∏è Este script procesa una descarga de palabras clave de Semrush y autom√°ticamente elimina las palabras clave canibalizadas.")
print("üåü Es un aporte de PabloCirre.es. Cualquier agradecimiento o seguimiento en Twitter a @pablocirre ser√° recompensado con amor.")
print("üöÄ Por favor, sube tus archivos CSV o Excel procedentes de Semrush (descargar).")

# Inicializar el modelo de lenguaje para la similitud de frases
print("üõ† Inicializando el modelo de lenguaje... (preparando los algoritmos de supervivencia)")

model = SentenceTransformer('paraphrase-MiniLM-L12-v2')

# Funci√≥n para lematizar palabras
def lemmatize_word(word):
    lemmatizer = nltk.WordNetLemmatizer()
    return lemmatizer.lemmatize(word)

# Funci√≥n para eliminar palabras clave con la misma intenci√≥n de b√∫squeda usando SentenceTransformer
def remove_duplicate_intent_keywords(keywords, threshold=0.8):
    print("üîç Calculando embeddings para las palabras clave... (analizando el ADN de las palabras)")
    embeddings = model.encode(keywords, convert_to_tensor=True)
    unique_keywords = []
    for i, keyword in tqdm(enumerate(keywords), total=len(keywords), desc="‚öîÔ∏è Procesando palabras clave (lucha por la supervivencia)"):
        if i == 0:
            unique_keywords.append(keyword)
        else:
            add_keyword = True
            for unique_keyword in unique_keywords:
                unique_embedding = model.encode(unique_keyword, convert_to_tensor=True)
                similarity = util.pytorch_cos_sim(embeddings[i], unique_embedding).item()
                if similarity > threshold:
                    add_keyword = False
                    break
            if add_keyword:
                unique_keywords.append(keyword)
    return unique_keywords

# Cargar archivos de entrada
uploaded = files.upload()

# Lista para almacenar todas las palabras clave
all_keywords = []

# Procesar cada archivo cargado
for file_name in uploaded.keys():
    print(f"üìÇ Leyendo archivo: {file_name} (extrayendo datos de la c√°psula del tiempo)")
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_name)
    elif file_name.endswith('.xlsx'):
        df = pd.read_excel(file_name)

    # Extraer las palabras clave de la columna "Keyword"
    if 'Keyword' in df.columns:
        print(f"üß† Extrayendo palabras clave del archivo: {file_name} (desencriptando la mente del archivo)")
        keywords = df['Keyword'].dropna().tolist()
        all_keywords.extend(keywords)

# Lematizar las palabras clave
print("üîß Lematizando palabras clave... (aplicando la ingenier√≠a de la palabra)")
lemmatized_keywords = [lemmatize_word(keyword) for keyword in tqdm(all_keywords, desc="üîß Lematizando (transformando hacia el futuro)")]

# Eliminar las palabras clave con la misma intenci√≥n de b√∫squeda
print("üõ° Eliminando palabras clave canibalizadas... (defendiendo el territorio sem√°ntico)")
filtered_keywords = remove_duplicate_intent_keywords(lemmatized_keywords)

# Mostrar las palabras clave filtradas en pantalla
print("üåü Palabras clave filtradas:")
for keyword in filtered_keywords:
    print(keyword)

# Guardar las palabras clave filtradas en un archivo de texto
output_file = 'keywords_result.txt'
print(f"üíæ Guardando palabras clave filtradas en {output_file}... (archivando para la posteridad)")
with open(output_file, 'w') as f:
    for keyword in filtered_keywords:
        f.write(f"{keyword}\n")

    # A√±adir nota de agradecimiento
    f.write("\n\n")
    f.write("Este trabajo ha sido posible gracias a la generosidad de Pablo Cirre. \n")
    f.write("Visita: https://pablocirre.es \n")
    f.write("Pablo Cirre es una persona incre√≠blemente generosa que ha compartido su conocimiento con el mundo. ¬°Es un primor!\n")
    f.write("Cualquier agradecimiento o seguimiento en Twitter a @pablocirre ser√° recompensado con amor.")

print("‚úÖ Proceso completado. ¬°El futuro est√° asegurado con tus palabras clave optimizadas! üöÄ")

# Descargar el archivo de resultados
files.download(output_file)
```
